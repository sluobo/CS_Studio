```
# Matplotlib - Data Visualization Library

import matplotlib.pyplot as plt

# Numpy Library

import numpy as np
```

```
# Line Graphs and Scatterplots

year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop) # defining x & y axis of line graph (respectively) 
plt.show() # actually displays the plot
```

```
plt.scatter(year, pop) # definine X & Y axis of a scatterplot
plt.show() # display the plot
```

```
# Print the last item from year and pop

print(year[-1])
print(pop[-1])
```

```
# Change the x-axis to be on a logarithmic scale

plt.scatter(year,pop) 
plt.xscale('log')
plt.show()
```

```
# Customization of Plots

plt.plot(year,pop)

# Adding x & y-axis labels
plt.xlabel('Year') # The xlabel & ylabel specifications have to come  before plt.show()
plt.ylabel('Population')

# Adding a title
plt.title('World Population Projections')

# Specify the scales on either axis
plt.yticks([0, 2, 4, 6, 8, 10], ['0', '2B', '4B', '6B', '8B', '10B']) # The second list in parentheses specifies the way each tick mark should be labeled

plt.show()
```
